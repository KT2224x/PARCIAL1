```python
def limpieza_tildes(df, rename_dict={}, do_inplace=True):
    if not rename_dict:
        df.rename(columns={col: col.lower().replace('á', 'a').replace('é', 'e').replace('í', 'i').replace('ó', 'o').replace('ú', 'u') for col in df.columns}, inplace=do_inplace)
    else:
        df.rename(columns={col: col.lower().replace('á', 'a').replace('é', 'e').replace('í', 'i').replace('ó', 'o').replace('ú', 'u') for col in df.columns}, inplace=do_inplace)
        df.rename(columns=rename_dict, inplace=do_inplace)
```

```python
limpieza_tildes(df,rename_dict={},do_inplace=True)
df.head()
```

```python
def limpieza_nombre_de_columnas(df,rename_dict={},do_inplace=True):
  if not rename_dict:
    return df.rename(columns={col: col.lower().replace(' ','_') for col in df.columns.values.tolist()}, inplace=do_inplace)
  else:
    return df.rename(columns=rename_dict,inplace=do_inplace)
```

```python
limpieza_nombre_de_columnas(df,rename_dict={},do_inplace=True)
df.head()
```

```python
df["fecha_de_muerte"].fillna("N/A", inplace=True)
df["codigo_iso_del_pais"].fillna("COL", inplace=True)
df["nombre_del_pais"].fillna("Colombia", inplace=True)
df["nombre_del_grupo_etnico"].fillna("N/P", inplace=True)
df["fecha_de_recuperacion"].fillna("N/A", inplace=True)
df["tipo_de_recuperacion"].fillna("N/A", inplace=True)
df.head()
```

```python
conteo_recuperados = df["recuperado"].value_counts()
plt.figure(figsize=(10, 10))
plt.bar(conteo_recuperados.index, conteo_recuperados.values, color=['green', 'red'])
plt.xlabel('recuperacion')
plt.ylabel('Número de Casos')
plt.title('Comparación de Casos recuperados')
plt.show()
```
